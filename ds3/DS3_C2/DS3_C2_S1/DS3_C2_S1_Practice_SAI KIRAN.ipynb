```python
t4i=[]
for i in range(len(test1)):
    if re.search('phenomena.*',test1[i].lower()):
        t4i.append(i)
t4 = df.iloc[t4i]
print('Dates of reviews with the word phenomena')
t4.date
```

```python
l42 = []
for i in t4['text']:
    l42.append(re.findall('\d+',i))
s42 = []
for i in l42:
    k = 0
    for j in i:
        k=k+float(j)
    s42.append(k)
sum42 = 0
for i in s42:
    sum42=sum42+i
average = sum42/len(s42) 
average
```

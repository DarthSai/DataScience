```python
#task 4
w1=[3,6,4,6,2,3,1,5,1,0,3,3]
w2=[1,2,4,0,2,6,5,4,2,5,3,4]
w3=[5,3,5,3,5,4,7,3,4,8,3,1]
w=[]
for i in range(0,12):
    a=(w1[i]+w2[i]+w3[i])/3
    w.append(a)
u=np.mean(w)
n=np.arange(0,21)
cust=poisson.pmf(n,u)
#a
print('a.The probability that no customers arrive in a 5 minute window is',cust[0])
#b
print('b.The probability that 6 or more customers arrive in a 5-minute interval is',1-cust[0:6].sum())
#c
u2=2*u
cust2=poisson.pmf(n,u2)
print('c.The probability that less than 4 customers arrive in a 10-minute interval is',cust2[0:4].sum())
#d
print('d.The probability that between 3 and 6 customers arrive in a 10 minute interval is',cust2[4:8].sum())
#e
u3=3*u
cust3=poisson.pmf(n,u3)
print('e.The possibility that exactly 8 customers arrive in a 15-minute intervaal is',cust3[9])
#f
fig,(c1,c2,c3)=plt.subplots(3,1)
c1.scatter(n,cust)
c1.set_title('5-minute interval')
c2.scatter(n,cust2)
c2.set_title('10-minute interval')
c3.scatter(n,cust3)
c3.set_title('15-minute interval')
plt.subplots_adjust(hspace=0.8)
```

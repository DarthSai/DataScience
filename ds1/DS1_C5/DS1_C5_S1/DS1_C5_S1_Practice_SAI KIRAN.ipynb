```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
cars=pd.read_csv('DS1_C5_S1_Descriptive_Analysis_Practice_Car_Data.csv')
cars.columns
```

```python
#task 1.1 and task 1.2
carmakemean=cars.groupby(by='Make')['Displacement','Fuel_Tank_Capacity_litre','City_Mileage_km_litre','Highway_Mileage_km_litre'].mean()
carmakemean
carmakemedian=cars.groupby(by='Make')['Displacement','Fuel_Tank_Capacity_litre','City_Mileage_km_litre','Highway_Mileage_km_litre'].median()
carmakemedian
carmakesd=cars.groupby(by='Make')['Displacement','Fuel_Tank_Capacity_litre','City_Mileage_km_litre','Highway_Mileage_km_litre'].std()
carmakesd
```

```python
#task 1.2(continued)
t11=pd.concat([carmakemean,carmakemedian,carmakesd],axis=1)
t11.columns=['Mean Displacement', 'Mean Fuel_Tank_Capacity_litre', 'Mean City_Mileage_km_litre',
       'Mean Highway_Mileage_km_litre', 'Median Displacement', 'Median Fuel_Tank_Capacity_litre',
       'Median City_Mileage_km_litre', 'Median Highway_Mileage_km_litre', 'Std.Dev Displacement',
       'Std.Dev Fuel_Tank_Capacity_litre', 'Std.Dev City_Mileage_km_litre',
       'Std.Dev Highway_Mileage_km_litre']
t11
t11['Coefficient of Variance of Displacement']=t11['Std.Dev Displacement']/t11['Mean Displacement']
t11['Coefficient of Variance of Fuel_Tank_Capacity_litre']=t11['Std.Dev Fuel_Tank_Capacity_litre']/t11['Mean Fuel_Tank_Capacity_litre']
t11['Coefficient of Variance of City_Mileage_km_litre']=t11['Std.Dev City_Mileage_km_litre']/t11['Mean City_Mileage_km_litre']
t11['Coefficient of Variance of Highway_Mileage_km_litre']=t11['Std.Dev Highway_Mileage_km_litre']/t11['Mean Highway_Mileage_km_litre']
t11
#Bugatti,Rolls-Royce,Aston Martin,Ferrari build cars that have the largest engines.
#Most cars in Aston Martin,Bentley,Ferrari,Lamborghini,Land Rover,Isuzu,Lexus,Maserati,Porsche have large fuel tanks.
#Datsun,Force,Honda,Hyundai,Lexus,Mahindra,Maruti Suzuki,Tata and Volkswagen have some cars that have some of thye highest mileages.
```

```python
#task 1.3
dispSuzuki=cars[(cars['Make']=='Maruti Suzuki')|(cars['Make']=='Maruti Suzuki R')]['Displacement'].tolist()
dispToyota=cars[cars['Make']=='Toyota']['Displacement'].tolist()
dispMahindra=cars[cars['Make']=='Mahindra']['Displacement'].dropna()
dispMahindra=dispMahindra.tolist()
plt.boxplot([dispSuzuki,dispToyota,dispMahindra],labels=['Suzuki','Toyota','Mahindra'])
#  Most cars in Suzuki fall in 1000cc to 1200cc with some extremes being 800 and 1500cc 
#  Toyota have a wider range of displacement whose least numbers sit atop Maruti Suzuki's higher side with some 
#- cars' engine at 3000cc.
#  While Mahindra and Toyota share some numbers at the lower end of displacement Mahindra has a narrow range
#- compared to Toyota but slightly more engine diplacement at the higher end than Maruti Suzuki
```

```python
#task 3.1.4.year
grouped_game=game.groupby(['Year'])
year=game.Year.unique()
Strat_Sample_game=pd.DataFrame()
for d in year:
    d_data=grouped_game.get_group(d) 
    sample_data=d_data.sample(frac=0.05,replace=True, random_state=1) 
    Strat_Sample_game=Strat_Sample_game.append(sample_data, ignore_index=True)
Strat_Sample_game
```

```python
#task 3.1.4.genre
grouped_game=game.groupby(['Genre'])
genre=game.Genre.unique()
Strat_Sample_gameg=pd.DataFrame()
for d in genre:
    d_data=grouped_game.get_group(d) 
    sample_data=d_data.sample(frac=0.05,replace=True, random_state=1) 
    Strat_Sample_gameg=Strat_Sample_gameg.append(sample_data, ignore_index=True)
Strat_Sample_gameg
```

```python
#task 3.1.5.year
grouped_games=game.groupby(['Year'])
years=[2018,2020]
games_Sample_df=pd.DataFrame()
for year in years:
    year_data=grouped_games.get_group(year) 
    sample_data=year_data.sample(frac=.2, random_state=1) 
    games_Sample_df=games_Sample_df.append(sample_data, ignore_index=True) 
games_Sample_df
```

```python
#task 3.1.5.genre
grouped_gamess=game.groupby(['Genre'])
genres=['Action','Racing']
gamess_Sample_df=pd.DataFrame()
for genre in genres:
    genre_data=grouped_gamess.get_group(genre) 
    sample_data=genre_data.sample(frac=.2, random_state=1) 
    gamess_Sample_df=gamess_Sample_df.append(sample_data, ignore_index=True) 
gamess_Sample_df
```

```python
#task 3.2.1
sample_10=game.sample(n=10, random_state=1)
sample_20=game.sample(n=20, random_state=1)
sample_30=game.sample(n=30, random_state=1)
sample_40=game.sample(n=40, random_state=1)
sample=[sample_10,sample_20,sample_30,sample_40]
agg=pd.DataFrame()
gmean=[]
gmedian=[]
gstd=[]
for i in sample:
    gmeang=i['Global_Sales_M$'].mean()
    gmean.append(gmeang)
    gmediang=i['Global_Sales_M$'].median()
    gmedian.append(gmediang)
    gstdg=i['Global_Sales_M$'].std()
    gstd.append(gstdg)

gmeans=pd.Series(gmean)
gmedians=pd.Series(gmedian)  
gstds=pd.Series(gstd)
t321=pd.DataFrame()
t321['Sample Size']=[10,20,30,40]
t321=t321.set_index('Sample Size')
t321['Mean']=[sample_10['Global_Sales_M$'].mean(),sample_20['Global_Sales_M$'].mean(),sample_30['Global_Sales_M$'].mean(),sample_40['Global_Sales_M$'].mean()]
t321['Median']=[sample_10['Global_Sales_M$'].median(),sample_20['Global_Sales_M$'].median(),sample_30['Global_Sales_M$'].median(),sample_40['Global_Sales_M$'].median()]
t321['Standard Deviation']=[sample_10['Global_Sales_M$'].std(),sample_20['Global_Sales_M$'].std(),sample_30['Global_Sales_M$'].std(),sample_40['Global_Sales_M$'].std()]
t321
```

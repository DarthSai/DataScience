```python
r='r'
nr='nr'
#function for calculating number of permutations:for repeating re=r or re='r',for non-repeating re=nr or re='nr'
def p(n,r,re):
    def fac(n):
        i=1
        f=1
        while i<=n:
            f=i*f
            i=i+1
        return f
    if re=='r':
        a=n**r
    elif re=='nr':
        a=fac(n)/fac(n-r)
    return a

#function for calculating number of combinations:for repeating re=r or re='r',for non-repeating re=nr or re='nr'
def c(n,r,re):
    def fac(n):
        i=1
        f=1
        while i<=n:
            f=i*f
            i=i+1
        return f
    if re=='r':
        a=fac(n+r-1)/(fac(r)*fac(n-1))
    elif re=='nr':
        a=fac(n)/(fac(r)*fac(n-r))
    return a
            
        
    
```

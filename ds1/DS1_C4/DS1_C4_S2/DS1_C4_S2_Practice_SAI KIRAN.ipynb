```python
import pandas as pd
import numpy as np
```

```python
#task 1
students=pd.Series(index=np.arange(1001,1041))
```

```python
#task 2
mathematicsl1=pd.Series(np.random.randint(50,95,len(students)),index=students.index,name='mathl1')
```

```python
#task 3
physicsl1=pd.Series(np.random.randint(40,95,len(students)),index=students.index,name='phyl1')
```

```python
#task 4
l1total=mathematicsl1+physicsl1
l1total
l1dic={'Roll':students.index,'MathL1':mathematicsl1,'PhysicsL1':physicsl1,'L1Total':l1total}
l1=pd.DataFrame(l1dic)
l1


```

```python
#task 5
l2=l1[l1total>=150]
l2
mathematicsl2=pd.Series(np.random.randint(50,100,len(l2)),name='mathl2',index=l2.index)
physicsl2=pd.Series(np.random.randint(30,100,len(l2)),name='phyl2',index=l2.index)
l2total=(mathematicsl2+physicsl2)
total=l2['L1Total']+l2total
l2=pd.concat([l2,mathematicsl2,physicsl2,l2total.rename('L2Total'),total.rename('Total')],axis=1)
l2.rename(columns={'0':'L2Total','1':'Total'})
l2
```

```python
#task 6
st=l2[total>=300]
st
```

```python
#task 7
mathematicsl3=pd.Series(np.random.randint(50,100,len(st)),name='mathl3',index=st.index)
physicsl3=pd.Series(np.random.randint(50,100,len(st)),name='phyl3',index=st.index)
l3total=mathematicsl3+physicsl3
evalscr=st['Total']*40/400+l3total*60/200
st=pd.concat([st,mathematicsl3,physicsl3,l3total.rename('L3Total'),evalscr.rename('Final Percentage')],axis=1)
st
```

```python
#task 8
final=st[evalscr>=75]
final

```

```python
#task 9
maximum=final['Final Percentage'].idxmax()
print(maximum,"scored the highest score of",final['Final Percentage'].max())

print('')

minimum=final['Final Percentage'].idxmin()
print(minimum,"scored the lowest score of",final['Final Percentage'].min())
```

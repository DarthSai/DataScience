```python
import numpy as np
import pandas as pd
```

```python
bank=pd.read_csv('DS1_C4_S3_Loan_Data_Practice.csv')
T0=bank
bank
```

```python
#task 1
t1=bank[(bank['ApplicantIncome']<6000)&(bank['LoanAmount']>200)&(bank['Loan_Status']=='Y')]
print('There are',len(t1[t1['Gender']=='Male']),'males and',len(t1[t1['Gender']=='Female']),'females whose income is less than $6000 and loan amount is more than $200,000 whose loans are approved')
T1=t1[['Loan_ID','Gender']]
T1
```

```python
#task 2
surb=len(bank[bank['Property_Area']=='Semiurban'])
urb=len(bank[bank['Property_Area']=='Urban'])
print('There are',surb,'customers who have property in semi-urban areas and',urb,'customers who have property in urban areas. The percentage of customers with property in semiurban areas is',round((surb-urb)/urb*100,2),'% higher than those who have property in urban areas.' )
T2_1=bank[bank['Property_Area']=='Semiurban']
T2_2=bank[bank['Property_Area']=='Urban']
```

```python
#task 3
bankincome=bank[(bank['Loan_Status']=='Y')].sort_values('ApplicantIncome',ascending=False)
T3=bankincome.iloc[:5]
T3
```

```python
#task 4
fem2dep=bank[(bank['Gender']=='Female')&(bank['Dependents']=='2')]
T4=fem2dep[['ApplicantIncome','LoanAmount']]
T4
```

```python
#task 5
bank.loc[((bank['Gender']=='Female')&(bank['LoanAmount']>200)),'Loan_Amount_Term']=180
T5=bank[(bank['Gender']=='Female')&(bank['LoanAmount']>200)]['Loan_Amount_Term']
T5
```

```python
#task 6
bank=bank.rename(columns={'Loan_Amount_Term':'LoanAmountTerm'})
bank.loc[12,'LoanAmountTerm']=60
bank.loc[36,'LoanAmountTerm']=60
T6=bank.loc[[12,36]]
T6
```

```python
#task 7
creditcard=bank[(bank['Education']=='Graduate')&(bank['Self_Employed']=='Yes')&(bank['ApplicantIncome']>10000)]
creditcard['Credit_Limit']=creditcard['ApplicantIncome']*2
T7=creditcard
T7
```

```python
#task 8
bankbackup=bank.copy()
kickout=bank[(bank['Dependents']=='3+')&(bank['Credit_History']==0)&(bank['Education']=='Not Graduate')]
ki=kickout.index
kickoutcopy=kickout.copy()
print(kickoutcopy)
bank.drop(ki,inplace=True)

T8_1=kickoutcopy
T8_2=bank

bank

```

```python
#task 9
grad=bank[bank['Education']=='Graduate']

notgrad=bank[bank['Education']=='Not Graduate']
print(grad['ApplicantIncome'],notgrad['ApplicantIncome'],end='    ')
T9_1=grad
T9_2=notgrad
```

```python
#task 10
import xlsxwriter
with pd.ExcelWriter('Bank Challenge2.xlsx',engine='xlsxwriter') as wr:
    T0.to_excel(wr,sheet_name='task0')
    T1.to_excel(wr,sheet_name='task1')
    T2_1.to_excel(wr,sheet_name='task2.1')
    T2_2.to_excel(wr,sheet_name='task2.2')
    T3.to_excel(wr,sheet_name='task3')
    T4.to_excel(wr,sheet_name='task4')
    T5.to_excel(wr,sheet_name='task5')
    T6.to_excel(wr,sheet_name='task6')
    T7.to_excel(wr,sheet_name='task7')
    T8_1.to_excel(wr,sheet_name='task8.1')
    T8_2.to_excel(wr,sheet_name='task8.2')
    T9_1.to_excel(wr,sheet_name='task9.1')
    T9_2.to_excel(wr,sheet_name='task9.2')
    
```

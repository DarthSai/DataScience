```python
import numpy as np
from numpy import genfromtxt
emp=genfromtxt("DS1_C4_S1_Employee_Data_Challenge.csv",delimiter=',',skip_header=1)
emp
```

```python
#task 1
averageage=np.mean(emp[:,1])
print(averageage)
```

```python
#task 2
agemorethan25=np.where(emp[:,1]>25)
print(emp[agemorethan25][:,0])
len(emp[agemorethan25])
```

```python
#task 3
agebetween30and35=np.where((emp[:,1]>=30) & (emp[:,1]<=35))
print(emp[agebetween30and35][:,0])
```

```python
shop=genfromtxt("DS1_C4_S1_Shopping_Data_Challenge.csv",delimiter=',',skip_header=1)
```

```python
#task 1
ssmorethan80=np.where(shop[:,3]>80)
print(shop[ssmorethan80])
```

```python
#task 2
agebetween20and24=np.where((shop[:,1]>=20)&(shop[:,1]<=24))
print(shop[agebetween20and24])
```

```python
#task 3
lowestincome=np.where(shop[:,2]==np.min(shop[:,2]))
print("Customers with lowest income")
print(shop[lowestincome])
print('')
highestincome=np.where(shop[:,2]==np.max(shop[:,2]))
print("Customers with highest income")
print(shop[highestincome])
```

```python
#task 4
disc=np.where((shop[:,3]>75) & (shop[:,2]>35) & (shop[:,1]<45))
print(shop[disc])
len(shop)
```

```python
#task 5
discount=np.array(np.random.randint(10,50,len(shop)))
shop=np.insert(shop,4,discount,axis=1)
shop
```

```python
#task 6
discountmorethan45=np.where(shop[:,4]>45)
print("Customers with discount more than 45%")
print(shop[discountmorethan45])
```

```python
#task 1  
hr=np.arange(0,500)
finance=np.arange(500,1000)
sales=np.arange(1000,1500)
it=np.arange(1500,2000)
print(hr)
print(finance)
print(sales)
print(it)
```

```python
#task 2
hrsalary=np.random.randint(25000,50000,500)
financesalary=1.25*hrsalary
salessalary=0.5*financesalary
itsalary=5000+hrsalary
print("HR")
print(hrsalary)
print("Finance")
print(financesalary)
print("Sales")
print(salessalary)
print("IT")
print(itsalary)
```

```python
#task 3
hrtotalctc=np.sum(hrsalary)
financetotalctc=np.sum(financesalary)
salestotalctc=np.sum(salessalary)
ittotalctc=np.sum(itsalary)
print("HR department total spend is",hrtotalctc)
print("Finance department total spend is",financetotalctc)
print("Sales department total spend is",salestotalctc)
print("IT department total spend is",ittotalctc)
```

```python
#task 4
hrpaysort=np.sort(hrsalary)
hrpaysortdesc=hrpaysort[::-1]
print("Top 10 earnings in HR")
print(hrpaysortdesc[0:10])
financepaysort=np.sort(financesalary)
financepaysortdesc=financepaysort[::-1]
print("Top 10 earnings in Finance")
print(financepaysortdesc[0:10])
```

```python
#task 5
allsalary=np.concatenate((hrsalary,financesalary,salessalary,itsalary))
allsalarymorethan45k=np.where(allsalary>45000)
allsalary[allsalarymorethan45k]*1.1
print(allsalary)

salespay=allsalary[1000:1500]
itpay=allsalary[1500:2000]
salespaymorethan45k=np.where(salespay>45000)
itpaymorethan45k=np.where(itpay>45000)
print("There are",len(salespay[salespaymorethan45k])," employees in Sales department and",len(itpay[itpaymorethan45k]),"employees in IT department whom earn more than $45000")
```

```python
#task 6
salespaylessthan45k=np.where(salespay<45000)
salespay1=salespay[salespaylessthan45k]*1.1
print(salespay1)
itpay1=allsalary[1500:2000]
hrpay1=allsalary[0:500]
financepay1=allsalary[500:1000]
newpay=np.concatenate((hrpay1,financepay1,salespay1,itpay1))
print(newpay)
```

```python
import numpy as np
from numpy import genfromtxt
college=genfromtxt("DS1_C4_S1_AdmissionPredict_Data_Practice.csv",delimiter=',',skip_header=1)
college
```

```python
gre=np.where(college[:,1]>320)
print(college[gre,:])
print("There are",np.size(gre),"students who scored above 320 in GRE")
```

```python
toefl=np.where(college[:,2]>110)
print("There are",np.size(toefl),"students who scored above 110 in TOEFL")
```

```python

scholarship=np.where((college[:,7]==1) & (college[:,1]>315) & (college[:,2]>115))
print(college[scholarship][:,0])
print("There are",len(college[scholarship]),"students who are eligible for scholarship for students with research experience and whose GRE score is greater   than 315 and TOEFL score is greater than 115.")
print(college[scholarship])
```

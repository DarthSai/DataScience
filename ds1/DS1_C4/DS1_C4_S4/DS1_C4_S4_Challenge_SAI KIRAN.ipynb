```python
import pandas as pd
import numpy as np
```

```python
cars=pd.read_csv('DS1_C4_S4_Car_Prices_Data_Challenge.csv')
cars
```

```python
#task 1
backupcars=cars.copy()
t1=cars[cars['year']==2014].sort_values('sellingprice',ascending=True).head(5)
t1[['make','model']]
```

```python
#task 2
t2=pd.pivot_table(cars[(cars['make']=='BMW')&(cars['condition']>4.5)],index=['year'],values=['sellingprice'],aggfunc='mean')
bmw2010avg=t2.loc[2010,'sellingprice']
bmw2015avg=t2.loc[2015,'sellingprice']
print('The average price of BMW cars in 2010 is',bmw2010avg,'and the average price in 2015 is',bmw2015avg,'which is',round((bmw2015avg-bmw2010avg)/bmw2010avg*100,2),'  percent higher than the price in 2010')
```

```python
#task 3
t3=pd.pivot_table(cars[(cars['interior']=='black')|(cars['interior']=='gray')],index=['interior'],values=['mmr'],aggfunc='mean')
blackavg=t3.loc['black','mmr']
grayavg=t3.loc['gray','mmr']
print('The price of cars with black interior is',blackavg,'and the price of cars with gray interior is',grayavg,'which is',round((blackavg-grayavg)/blackavg*100,2),'percent lower than cars with black interior')
```

```python
#task 4
t41=pd.crosstab(cars[cars.year.between(2013,2015)].make,cars[cars.year.between(2013,2015)].color)
t42=pd.crosstab(cars[cars.year.between(2013,2015)].color,cars[cars.year.between(2013,2015)].make)

t43=t42.idxmax()
t43
t44=t42.max()
maxcolor={'bestsellingcolor':t43,'carssoldforcolor':t44}
maxc=pd.DataFrame(maxcolor)
print(maxc)
t41
```

```python
#task 5
t5=pd.pivot_table(cars[(cars['body']=='SUV')&((cars['saledate']=='Fri Dec 19 2014 09:00:00 GMT-0800 (PST)')|(cars['saledate']=='Thu Jan 15 2015 04:30:00 GMT-0800 (PST)'))],index=['saledate'],values=['sellingprice'],aggfunc='sum')
dec19=t5.loc['Fri Dec 19 2014 09:00:00 GMT-0800 (PST)','sellingprice']
jan15=t5.loc['Thu Jan 15 2015 04:30:00 GMT-0800 (PST)','sellingprice']
print(t5)
print('The total sale for SUVs on Fri Dec 19 2014 09:00:00 GMT-0800 (PST) is',dec19,'and on Thu Jan 15 2015 04:30:00 GMT-0800 (PST) is',jan15,'which is',round((dec19-jan15)/dec19*100,2),'percent lower than on Fri Dec 19 2014 09:00:00 GMT-0800 (PST)')
```

```python
#task 6
discars=cars.copy()
discars.insert(15,'discountedprice',discars.sellingprice)
discars
discars.loc[((discars['make']=='Ford')|(discars['make']=='Chevrolet')),'discountedprice']=0.9*discars['sellingprice']
discars.loc[((discars['make']!='Ford')&(discars['make']!='Chevrolet')),'discountedprice']=0.8*discars['sellingprice']
discars
```

```python
#task 7
cars=discars.copy()
cars
del cars['saledate']
del cars['vin']
cars.isnull().sum()
makemode=cars['make'].mode()
cars['make']=cars['make'].fillna('Ford')
cars['model']=cars['model'].fillna('Altima')
cars['trim']=cars['trim'].fillna('Base')
cars['body']=cars['body'].fillna('Sedan')
cars['transmission']=cars['transmission'].fillna('automatic')
cars['state']=cars['state'].fillna('fl')
cars['condition']=cars['condition'].fillna(cars['condition'].median())
cars['odometer']=cars['odometer'].fillna(cars['odometer'].mean())
cars['color']=cars['color'].fillna('black')
cars['interior']=cars['interior'].fillna('black')

cars.isnull().sum()
```

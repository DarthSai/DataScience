```python
#displays the inventory to the customer
inventory={1:["Biscuits",5,20.50],2:["Cereal",10,90.00],3:["Chicken",20,200.00],4:["Oats",20,39.99],5:["Rice",30,79.99]}
print("       ESSENTIAL ITEMS AVAILABLE")
print("{:10} {:10} {:10} {:10}".format('S.no','Item','Quantity','Cost/Item'))
for (Sno, [Item,Quantity,Costperitem]) in inventory.items():
    print("{:<10} {:<10} {:<10} {:<10}".format(Sno,Item,Quantity,Costperitem))

#takes input of quantity of items required by the customer and checks if it's available in the inventory
bis=int(input("How many biscuit packets are required ?        "))
if bis>inventory[1][1]:
    print("We do not have sufficient Biscuits in stock,we only have",inventory[1][1],inventory[1][0],"in stock.Please restart your order and modify your order    accordingly")
cer=int(input("How many cereal packets are required ?         "))
if cer>inventory[2][1]:
    print("We do not have sufficient Cereals in stock,we only have",inventory[2][1],inventory[2][0],"in stock.Please restart your order and modify your order      accordingly")
chi=int(input("How many frozen chicken packets are required ? "))
if chi>inventory[3][1]:
    print("We do not have sufficient Chicken in stock,we only have",inventory[3][1],inventory[3][0],"in stock.Please restart your order and modify your order      accordingly")
oat=int(input("How many oats packets are required ?           "))
if oat>inventory[4][1]:
    print("We do not have sufficient Oats in stock,we only have",inventory[4][1],inventory[4][0],"in stock.Please restart your order and modify your order            accordingly")
ric=int(input("How many rice packets are required ?           "))
if ric>inventory[5][1]:
    print("We do not have sufficient Rice in stock,we only have",inventory[5][1],inventory[5][0],"in stock.Please restart your order and modify your order            accordingly")
order=[bis,cer,chi,oat,ric]
#used to calculate the prices of each item
prices=[]   
for i in range(0,len(inventory)):
    prices.append(inventory[i+1][2])
listsproduct=lambda l1,l2:list(map(lambda x, y: x*y, l1, l2)) #function to calculate subtotal price by type of item
subtotal=listsproduct(order,prices) #stores subtotals by item in a list

#takes input for customer's delivery details
print(" ")
print("Fill in your delivery details")

name=input("Enter your name: ")
address=input("Enter your address: ")
distore=int(input("Please enter the distance to our store from the address mentioned:"))
#calculate the delivery charge
if distore<=5:
    delch=10
elif distore<=10 and distore>5:
    delch=20
elif distore<=15 and distore>10:
    delch=30

if delch==10 or delch==20 or delch==30:
    print("Delivery charge of",delch,"will be levied for a distance of",distore,"kilometres")
else:
    print("We are sorry as we currently do not service your location")

#prints the output table for items seleected by customer along with subtotal by item, total for items and total including delivery charge  
print(" ")
print(" ")
print("         BILL")
print("{:10} {:10} {:10} {:10}".format('S.no','Item','Quantity','SubtotalItem'))
for i in range(0,5):
    print("{:<10} {:<10} {:<10} {:<10}".format(i+1,inventory[i+1][0],order[i],order[i]*inventory[i+1][2]))

print(" ")
print("Total Item Cost: ",sum(subtotal))
print(" ")
print("Total Bill Amount(includes delivery charge): ",sum(subtotal,delch))
print(" ")
print("Name: ",name)
print("Address: ",address)
print(" ")

#order confirmation
finalize=input("Do you want to finalize you order ?(y/n): ")
if finalize=='y':
    print("Thank you for shopping with us, your order will be delivered shortly.")
else:
    print("You have chosen to abandon your cart.")
    
print("Have a nice day!")
```
